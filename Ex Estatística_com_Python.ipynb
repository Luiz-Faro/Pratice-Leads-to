**3)	Temos a necessidade de calcular uma amostra estratificada a partir da base de dados de Diabetes do sklearn. Faremos uma amostragem estratificada por gênero, utilizando a variável “sex” da base, e depois selecionaremos uma fração aleatória de 5% de itens de cada grupo para a amostra. O rascunho do código se encontra abaixo, assinale a alternativa que completa os espaços vazios do código de forma a atingir o objetivo proposto acima.**

```python
diabetes_df = pd.DataFrame(datasets.load_diabetes().data)
diabetes_df.columns = datasets.load_diabetes().feature_names
amostra_diabetes = diabetes_df.groupby("sex").apply(lambda x: x.sample(frac=0.05, random_state=42))
```

```python
amostra_diabetes
```

```python
len(amostra_diabetes)
```

```python
len(diabetes_df)
```

```python
frac_luiz = (len(amostra_diabetes)/len(diabetes_df))*100
frac_luiz
```

**4)	Alguns docentes estão estudando as notas em um curso de Física. O mínimo que um estudante pode tirar é zero e o máximo 100. Os docentes observaram que as notas dos estudantes nesse curso seguem uma distribuição normal com média 50.2, conforme figura abaixo.**

```python

from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

data = np.linspace(10, 100, 100)

ax.plot(data, norm.pdf(data, loc=50.2, scale=8), label='mean=%.1f' % 50.2)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Notas de Estudantes em um Curso de Física')
ax.legend(loc='best', frameon=True)
ax.grid(True)
```

```python
probabilidade = norm.cdf(40, loc=50.2, scale=8) - norm.cdf(60, loc=50.2, scale=8)
probabilidade
```

```python
probabilidade_2 = norm.cdf(60, loc=50.2, scale=8) - norm.cdf(40, loc=50.2, scale=8)
probabilidade_2
```

```python
probabilidade_3 = (norm.pdf(60, loc=50.2, scale=8) - norm.pdf(40, loc=50.2, scale=8))*100
probabilidade_3
```

```python
probabilidade_4 = (norm.pdf(60, loc=50.2, scale=8) + norm.pdf(40, loc=50.2, scale=8))*100
probabilidade_4
```
